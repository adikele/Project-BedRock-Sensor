```python
#Problem Statement: 
#Data from an accelerometer sensor placed under a bed is available in HEX form. 
#The data represents displacement of the bed about a mean position.  
#The task is to (i) plot a displacement graph and:
#               (ii) to plot the rms values over specific time intervals so as to find intervals of low and high "activity"
                    #and (in subsequent analyses) to find the effect of noise on the signal
# Here "activity" denotes periods of high displacement signals.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
%pylab

# Tietoa voidaan lukea rivi kerrallaan.
file_name = 'A1_raaka_1.txt'
file = open(file_name,'r')

#.................



#::::::::::::
fs = 1000

N = 60*fs        # Segmentin pituus näytteinä, assuming that sensor is collecting 1000 samples per second 
t = arange(N)/fs # Aika (s) (OK, understood! arange (N) will create an array of 60 * 1000 values, so we have 60 000 values for t; we need
                 # them all but the highest value should be 60 (or close to 60) to indicate 60 s)
x = zeros(N)     # Alusta muuttuja x 
K = 10           # segmenttien määrä "when changed to K = 20, shows more lines, 
                 # line 14 has a much higher signal level than any in the first ten lines
                 # line 18 has a really high signal level (the highest)

figure(1) # Avaa uusi ikkuna
for seg in range(K):
    for i in range(N):
        # lue rivi ja muunna kokonaisluvuksi
        x[i] = int(file.readline(), 16)
        
    # Vähennä 567 ja jaa 100:lla
    x = (x - 567)/100.0
    
    # Piirrä segmentti ikkunaan
    fig2 = plt.figure(1)
    plot(t, x + seg*1.0, 'b') # I suppose that in executing the plot function once, python will fill one line of the display page
                              # That must be so, because when control returns to the plot fn in the next 'seg' iteration, it has a 
                              # plot with a 't' of 0 - 59.999 

# Lisää otsikot ja gridi
grid()
xlabel('Aika (s)')
ylabel('Amplitudi')
title('A1_raaka_1.txt')
show()

file.close()
#:::::::::::

file = open(file_name,'r')

fsf = 1000        # Näyteenottotaajuus Could it be that 1000 samples per minute?
timef = 3600      # NOT USED! Siirtymä (s)   60 min 
P = timef*fsf      # NOT USED! Siirtymä näytteinä time*fs  ..Take samples for 60 min?
Nf = int(5*P)     # NOT USED! Siirtymä tavuina 5*P .. ?? Translates to: Transition in bytes 5 * P
#pos = file.seek(Nf, 0)

lista = [[] for i in range(6)] #a list of 6 sublists
value = zeros (6) #array to store 6 rms values 

fs = 1000
N = 60*fs ##Phase 2.1: 60s = 1 min, the time over which one row's samples are being processed

#to calculate step: 
#step = (How many microseconds in 10 s?) Ans: 10 000

step = 10000 #step is the time interval over which the rms values are being sampled Phase 2.1:

t = arange(0, N, step)/fs #putting the timevalues from 0 to N-1 in an array, 
                           #so that 1000 values make 1 (sec)
                           #since i have the rms value for only 10 values at intervals of 30 which are 0, (30/300 - 1)s, ... so on, 
                           #arange fn has a step of 30 s
                           #each time values is .001s apart
            
#opening file and populating each of the sublists with 30 values each
for row in range(10): #creating a new row in the plot
    #for i in range(10): #Phase 1
    for i in range(6): #plotting six values in one row #Phase 2.1:
        sumx2 = 0
        for j in range (step): #for one time interval, process "step" number of time samples
            rivi = file.readline()   #read in one value 
            rivi_VainDigits = int(rivi[:-1], 16) #take off the tailing newline character 
            rivi_VainDigits = rivi_VainDigits - 567 #the raw value with respect to a central line
            rivi_VainDigits = rivi_VainDigits ** 2 #the square of the raw values
            lista [i].append (rivi_VainDigits) # append this value to the existing values in the current list
        
        # now "step" number of  values have been filled in the first sublist, lista [0]
        temp1 = sum (lista [i]) #finding the sum of the squared values 
        temp2 = (temp1 / step)  #finding the mean of the squared values 
        temp3 = sqrt (temp2)  #finding the root (of the mean of the squared values)
        value[i] = temp3 #for every i loop, one value of "value" will be added 
                         #for every row, the value array will be filled again     
        lista [i] = [0] * step
    
    print(value[i])#Phase 2.5 -- prints the rms value of the first interval for each of the rows. 
    fig2 = plt.figure(2)
    plot(t, value + row*1.0, 'b') #plots one row of "values"
    
file.close()

#so value has 10 values (same as t which has only 10 values between 0 and N, as they are in steps of 30 000)

#for plotting xvalues:
#note: we talked about having 30 second intervals, let's have 10 such intervals (so 5 min or 300s)
# Lisää otsikot ja gridi
grid()
xlabel('Aika (s)')
ylabel('rms')
title('A1_raaka_1.txt')
show()
```
